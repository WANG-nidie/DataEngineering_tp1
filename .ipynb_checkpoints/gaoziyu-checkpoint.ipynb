```python
import operator

count = 0
lis = []
lis_diff  =[]
for i in range(0,df.iloc[:,0].size):
    for j in range(0,df1.iloc[:,0].size):
        if df["姓名"][i] == df1["姓名"][j]:
           # print(str(df["姓名"][i])+"--"+str(df["学习总课时"][i])+"--"+str(df1["学习总课时"][j])+"--"+
                  #str(operator.eq(int(df["学习总课时"][i]),int(df1["学习总课时"][j])))) 
            if operator.eq(int(df["学习总课时"][i]),int(df1["学习总课时"][j])) == True:
                count +=1
                lis.append(df["姓名"][i])
            else:
                lis_diff.append(df["姓名"][i])
print("worksheet和sheet3课时相同的有"+str(count)+"个人，名单为:")
print(lis)
print("worksheet和sheet3课时不相同的名单为:")
print(lis_diff)
```
